```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
indep_hev_summary = pd.read_csv('indep/indep_limit_summary.csv')
indep_hev_summary
```

```python
pos_hev_summary = pd.read_csv('pos/pos_limit_summary.csv')
pos_hev_summary
```

```python
rep_correlation_summary = pd.concat([indep_hev_summary['offertimes'],
                                     indep_hev_summary['collection_size'],
                                     indep_hev_summary['indep_rep'],
                                     indep_hev_summary['indep_rep_runtime'],
                                     indep_hev_summary['indep_rep_runtime_se'],
                                     pos_hev_summary['pos_rep'],
                                     pos_hev_summary['pos_rep_runtime'],  
                                     pos_hev_summary['pos_rep_runtime_se']
                                     ],axis=1)
rep_correlation_summary.to_csv('coupla_repsentation.csv')
rep_correlation_summary
```

```python
limit_correlation_summary = pd.concat([indep_hev_summary['offertimes'],
                                     indep_hev_summary['collection_size'],
                                     indep_hev_summary['indep_avg_loss'],
                                     indep_hev_summary['indep_avg_loss_se'],
                                     indep_hev_summary['indep_limit_runtime'],
                                     indep_hev_summary['indep_limit_runtime_se'],
                                     pos_hev_summary['pos_avg_loss'],
                                     pos_hev_summary['pos_avg_loss_se'],
                                     pos_hev_summary['pos_limit_runtime'],
                                     pos_hev_summary['pos_limit_runtime_se']
                                     ],axis=1)
limit_correlation_summary.to_csv('coupla_limit.csv')
limit_correlation_summary
```
