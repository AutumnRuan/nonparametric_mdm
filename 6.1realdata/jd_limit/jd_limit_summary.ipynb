```python
limit_result = pd.concat([df_mdm['offertimes'].round(3),
                    df_mdm['mdm_loss'].round(3),
                    df_rum['rum_loss'].round(3),
                    df_mem['mem_loss'].round(3),
                    df_mnl['mnl_loss'].round(3),
                    df_mccm['mccm_loss'].round(3),
                    df_lcmnl['lcmnl_loss'].round(3),
                    df_indep['indep_loss'].round(3)
                    ], axis=1)
limit_result.to_csv('limit/total_loss_round3.csv')
limit_result
```

```python
total_offertimes = pd.read_csv('total_offertimes.csv')
total_offertimes = total_offertimes.drop(total_offertimes.columns[0],axis=1)
total_offertimes = total_offertimes.values.tolist()
total_offertimes_lst = []
for i in range(len(total_offertimes)):
    total_offertimes_lst.append(total_offertimes[i][0])
total_offertimes_lst
```

```python
avg_mdm_loss = []
avg_rum_loss = []
avg_mnl_loss = []
avg_mccm_loss = []
avg_lcmnl_loss = []
avg_mem_loss = []
for i in range(len(total_offertimes_lst)):
    avg_mdm_loss.append(df_mdm['mdm_loss'][i]/total_offertimes_lst[i])
    avg_mem_loss.append(df_mem['mem_loss'][i]/total_offertimes_lst[i])
    avg_mnl_loss.append(df_mnl['mnl_loss'][i]/total_offertimes_lst[i])
    avg_rum_loss.append(df_rum['rum_loss'][i]/total_offertimes_lst[i])
    avg_mccm_loss.append(df_mccm['mccm_loss'][i]/total_offertimes_lst[i])
    avg_lcmnl_loss.append(df_lcmnl['lcmnl_loss'][i]/total_offertimes_lst[i])
avg_limit_loss_summary = pd.DataFrame({'avg_mdm_loss':avg_mdm_loss,'mdm_time':df_mdm['mdm_limit_runtime'],
                                       'avg_rum_loss':avg_rum_loss,'rum_runtime':df_rum['rum_limit_runtime'],
                                       'avg_mem_loss':avg_mem_loss,'mem_time':df_mem['mem_limit_runtime'],
                                       'avg_mnl_loss':avg_mnl_loss,'mnl_time':df_mnl['mnl_limit_runtime'],
                                       'avg_mccm_loss':avg_mccm_loss,'mccm_runtime':df_mccm['mccm_limit_runtime'],
                                       'avg_lcmnl_loss':avg_lcmnl_loss,'lcmnl_runtime':df_lcmnl['lcmnl_limit_runtime'],
                                       'avg_indep_loss':df_indep['avg_indep_loss'],'indep_runtime':df_indep['indep_limit_runtime']})
avg_limit_loss_summary.to_csv('limit/avg_loss_summary.csv')
avg_limit_loss_summary
```

```python
avg_mdm_loss = []
avg_rum_loss = []
avg_mnl_loss = []
avg_mccm_loss = []
avg_lcmnl_loss = []
avg_mem_loss= []
for i in range(len(total_offertimes_lst)):
    avg_mdm_loss.append(df_mdm['mdm_loss'][i]/total_offertimes_lst[i]*1000)
    avg_rum_loss.append(df_rum['rum_loss'][i]/total_offertimes_lst[i]*1000)
    avg_mnl_loss.append(df_mnl['mnl_loss'][i]/total_offertimes_lst[i]*1000)
    avg_mccm_loss.append(df_mccm['mccm_loss'][i]/total_offertimes_lst[i]*1000)
    avg_lcmnl_loss.append(df_lcmnl['lcmnl_loss'][i]/total_offertimes_lst[i]*1000)
    avg_mem_loss.append(df_mem['mem_loss'][i]/total_offertimes_lst[i]*1000)
    
avg_limit_loss_summary = pd.DataFrame({'avg_mdm_loss':np.array(avg_mdm_loss).round(2),'mdm_time':df_mdm['mdm_limit_runtime'].round(2),
                                       'avg_rum_loss':np.array(avg_rum_loss).round(2),'rum_runtime':df_rum['rum_limit_runtime'].round(2),
                                       'avg_mem_loss':np.array(avg_mem_loss).round(2),'mem_time':df_mem['mem_limit_runtime'].round(2),
                                       'avg_mnl_loss':np.array(avg_mnl_loss).round(2),'mnl_time':df_mnl['mnl_limit_runtime'].round(2),
                                       'avg_mccm_loss':np.array(avg_mccm_loss).round(2),'mccm_runtime':df_mccm['mccm_limit_runtime'].round(2),
                                       'avg_lcmnl_loss':np.array(avg_lcmnl_loss).round(2),'lcmnl_runtime':df_lcmnl['lcmnl_limit_runtime'].round(2),
                                       'avg_indep_loss':df_indep['avg_indep_loss']*1000,'indep_runtime':df_indep['indep_limit_runtime']})
avg_limit_loss_summary.to_csv('limit/avg_loss_10-3_time.csv')
avg_limit_loss_summary
```
