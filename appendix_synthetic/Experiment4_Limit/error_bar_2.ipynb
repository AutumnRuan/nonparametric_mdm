```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_mnl = pd.read_csv('mnl_mle_summary.csv')
mnl_loss = np.array(df_mnl.loc[:,['mnl_mean_oss']])
mnl_se = np.array(df_mnl.loc[:,['mnl_loss_se']])
mnl_mean_loss = np.stack(mnl_loss,axis=1)
mnl_mean_loss = mnl_mean_loss[0]
mnl_loss_se = np.stack(mnl_se,axis=1)
mnl_loss_se = mnl_loss_se[0]
```

```python
df_mdm = pd.read_csv('mdm_limit_summary.csv')
mdm_loss = np.array(df_mdm.loc[:,['mdm_mean_oss']])
mdm_se = np.array(df_mdm.loc[:,['mdm_loss_se']])
mdm_mean_loss = np.stack(mdm_loss,axis=1)
mdm_mean_loss = mdm_mean_loss[0]
mdm_loss_se = np.stack(mdm_se,axis=1)
mdm_loss_se = mdm_loss_se[0]
```

```python
df_rum = pd.read_csv('rum_limit_summary.csv')
rum_loss = np.array(df_rum.loc[:,['rum_mean_oss']])
rum_se = np.array(df_rum.loc[:,['rum_loss_se']])
rum_mean_loss = np.stack(rum_loss,axis=1)
rum_mean_loss = rum_mean_loss[0]
rum_loss_se = np.stack(rum_se,axis=1)
rum_loss_se = rum_loss_se[0]
```

```python
collection_size = np.array([2,3,5,10,15])
```

```python
plt.errorbar(collection_size,mdm_mean_loss,yerr=mdm_loss_se,ecolor='r',elinewidth=2,capsize=4,label = 'mdm')
plt.errorbar(collection_size,rum_mean_loss,yerr=rum_loss_se,ecolor='r',elinewidth=2,capsize=4,label = 'rum')
plt.errorbar(collection_size,mnl_mean_loss,yerr=mnl_loss_se,ecolor='r',elinewidth=2,capsize=4,label = 'mnl')
#plt.errorbar(np.array([2,3,5,10]),apu_mean_loss,yerr=apu_loss_se,ecolor='r',elinewidth=2,capsize=4,label = 'apu')

y = [0,1,2,3,4]
plt.yticks(y,y)
plt.xticks(collection_size,collection_size)
plt.tick_params(labelsize=20)
plt.xlabel('collection size',fontsize = 20)
plt.ylabel('mean loss',fontsize = 20)
plt.legend(loc='upper left',fontsize = 20)

#plt.title(' limit loss comparison ')
plt.tight_layout()
plt.savefig('result/limit_loss_comparison_NEW.pdf')
plt.savefig('result/limit_loss_comparison_NEW.png')
plt.show()
```
