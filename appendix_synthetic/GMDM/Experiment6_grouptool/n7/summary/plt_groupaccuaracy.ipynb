```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('accuracy_summary.csv',header=0)
accuarcy = np.array(df.loc[:,['mean accuracy']])
se = np.array(df.loc[:,['std']])

mean_assortment_accuarcy = np.stack(accuarcy,axis=1)
mean_assortment_accuarcy = mean_assortment_accuarcy[0]
std_assortment_accararcy = np.stack(se,axis=1)
std_assortment_accararcy = std_assortment_accararcy[0]
```

```python
collection_size = np.array(df.loc[:,['assortment size']]).tolist()
collection_size = np.stack(collection_size,axis=1)[0]
collection_size
```

```python
########### plt the result ##############

''' visulize the prediction interval comparison '''
#plt.errorbar(collection_size,mdm_loss_mean,yerr=mdm_loss_std/np.sqrt(1000),marker='o',label = 'mdm')
#plt.errorbar(x,y,yerr=dy,fmt='o',ecolor='r',color='b',elinewidth=2,capsize=4)
plt.errorbar(collection_size,mean_assortment_accuarcy,yerr=std_assortment_accararcy,ecolor='r',marker='o',elinewidth=2,capsize=4)
#plt.errorbar(size,mdm_avg_len_interval,yerr=mdm_len_interval_se,ecolor='r',elinewidth=2,capsize=4,label = 'G-MDM without group')
y = [0.2,0.4,0.6,0.9,1.2]
plt.yticks(y,y)
plt.xticks(collection_size,collection_size)
plt.tick_params(labelsize=20)
plt.xlabel('collection size',fontsize = 20)
plt.ylabel('average accuracy',fontsize = 20)
plt.tight_layout()
#plt.legend(loc='upper right')
#plt.title('Prediction Accuracy with different sizes of assortment collections')
plt.savefig('group_accuracy.pdf')
plt.savefig('group_accuracy.png')
plt.show()
```
