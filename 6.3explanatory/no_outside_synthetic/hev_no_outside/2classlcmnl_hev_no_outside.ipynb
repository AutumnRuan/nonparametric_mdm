```python
def new_data_w_at_least_2_products(n,curr_collection,curr_choice_collection,curr_offertimes):
    
    pre_new_collection = []
    new_collection_indices = []
    new_offertimes = []
    
    for i in range(len(curr_collection)):
        if len(curr_collection[i])>2:
            pre_new_collection.append(curr_collection[i][1:])
            new_collection_indices.append(i)
            new_offertimes.append(curr_offertimes[i])
            
    # Reduce values of all elements by 1
    new_collection = [[element - 1 for element in sublist] for sublist in pre_new_collection]

    new_choice_collection = np.zeros((n-1,len(new_collection_indices)))
    for i in range(len(new_collection_indices)):
        new_choice_collection[:,i] = curr_choice_collection[1:,new_collection_indices[i]]/np.sum(curr_choice_collection[1:,new_collection_indices[i]])
            
    return new_collection,new_choice_collection,new_offertimes
```

```python
def collection_distribution_hev(n,collection):
    # generate mean 
    ''' mu_0 = np.random.uniform(2,3)
    mu_1n = np.random.uniform(-3,-2,n-1) '''
    # generate deterministic utilities for products
    
    ''' rho = -0.5 # prices and utilities are negatively correlated
    price_mean = np.mean(price[1:]) 
    price_std = np.std(price[1:])
    
    z = np.zeros(len(price[1:]))
    for i in range(len(z)):
        z[i] = (price[1:][i] - price_mean)/price_std
        
    mu_1n = np.zeros(len(z))
    normal_rvs = np.random.randn(len(mu_1n))
    for i in range(len(mu_1n)):
        mu_1n[i] = rho*z[i] + np.sqrt(1-rho**2)*normal_rvs[i] 
    ## deterministic utility of the outside option is strictly greater than the utilities of the products
    mu_0 = np.max(mu_1n) + 1
    # np.random.uniform(-2,2,n-1)
    mu = np.hstack((mu_0,mu_1n))
    
    scale_0 = 10
    scale_1n = np.random.uniform(0.04,1,n-1)
    scales = np.hstack((scale_0,scale_1n)) '''
    
    mu = np.random.uniform(-2,2,n)
    scales = np.random.uniform(0.04,2,n)
    
    sample_size = 10000
    utility_samples = np.array([gumbel_r.rvs(loc=mu, scale=scales) for _ in range(sample_size)])
    # generate covariance matrix with positive correlation
    # neg_cov = generate_negatively_correlated_covariance_matrix(n)
    # if is_positive_semidefinite(neg_cov): 
    #     sample_size = 10000
    #     utility_samples = np.random.multivariate_normal(mu,neg_cov,size=sample_size)
    # else:
    #     print('Negative correlation matrix error')
    
    collection_distribution = np.zeros((n,len(collection)))
    for i in range(len(collection)):
        curr_assortment = collection[i]
        curr_population = [[] for _ in range(sample_size)] 

        for j in range(sample_size):
            for k in curr_assortment:
                curr_population[j].append(utility_samples[j][k])
                # each sub list records only the utilities of products in the current assortment
                
        frequency = [0]*len(curr_assortment)
        for j in range(sample_size):
            max_index = np.argmax(np.array(curr_population[j]))
            # product is chosen iff the utility of the product is max in the assortment
            frequency[max_index] = frequency[max_index] +1 
            # update the frequency of product to be chosen
            
        prob = np.array(frequency)/np.sum(frequency)
        for j in range(len(curr_assortment)):
            collection_distribution[curr_assortment[j]][i] = prob[j]
            
    return collection_distribution
```

```python
import pyomo.environ as pyo
from pyomo.environ import *
from pyomo.opt import SolverFactory
```

```python
def expectation_step(v_values,data,collections,L,offer_times):
    
    n,m = data.shape

    # Pyomo model
    model = pyo.ConcreteModel()

    # Variables
    model.w = pyo.Var(range(L), domain=pyo.NonNegativeReals,bounds=(0.001,1))

    # Objective function
    model.obj = pyo.Objective(
        expr=sum(sum(offer_times[i] * data[j, i] * pyo.log(
            sum(model.w[l] * np.exp(v_values[l, j]) / sum(np.exp(v_values[l, k]) for k in collections[i]) for l in range(L))
        ) for j in collections[i]) for i in range(m)),
        sense=pyo.maximize
    )

    # Constraint: Sum of weights equals 1
    model.sum_weights_constraint = pyo.Constraint(expr=sum(model.w[l] for l in range(L)) == 1)

    # Solve the optimization problem
    solver = pyo.SolverFactory('ipopt')
    results = solver.solve(model)
    
    # Print optimization status
    #print("Optimization Status:", results.solver.status)
    if results.solver.status == pyo.SolverStatus.ok:
        #print('Optimal solution found!')
       
        #model.pprint()
    
        w_values = [model.w[l]() for l in range(L)]
        w_values = np.array(w_values)
        
        ''' print("Variable w =", [ model.w[l]() for l in range(L)])
        print("Objective Value =", model.obj()) '''
        
        return w_values,pyo.value(model.obj())
    else:
        print('No solution found or an error occurred.')
```

```python
def maximization_step(weights,data,collections,offer_times):
    # Given data
    L = len(weights)
    n,m = data.shape
    
    # Create a ConcreteModel
    model = pyo.ConcreteModel()

    # Define decision variables
    model.v = pyo.Var(range(L), range(n), bounds=(-100, 100))

    # Define the objective function
    model.obj = pyo.Objective(
        expr=sum(sum(data[j][i] * offer_times[i] * pyo.log(sum(weights[l] * pyo.exp(model.v[l, j]) / sum(pyo.exp(model.v[l, k]) for k in collections[i]) for l in range(L))) for j in collections[i]) for i in range(m)),
        sense=pyo.maximize
    )
    model.outsideopt_constraint = pyo.ConstraintList()
    for l in range(L):
        model.outsideopt_constraint.add(model.v[l,0] == 0)

    # Solve the model
    solver = pyo.SolverFactory('ipopt')
    results = solver.solve(model)

    # Print optimization status
    #print("Optimization Status:", results.solver.status)
    if results.solver.status == pyo.SolverStatus.ok:
        #print('Optimal solution found!')
        #model.write('lcmnl2_maximizatioin.mps', io_options={'symbolic_solver_labels': True})
        #model.pprint()
    
        v_values = [[model.v[l, i]() for i in range(n)] for l in range(L)]
        v_values = np.array(v_values)

        # Print variable values and objective value
        ''' print("Variable v =", [[model.v[l, i]() for i in range(n)] for l in range(L)])
        print("Objective Value =", model.obj()) '''
        
        return v_values,model.obj()
    else:
        print('No solution found or an error occurred.')
    
```

```python
# EM algorithm for solving the MLE of LC-MNL
def expectation_maximization(data, collections,offer_times, num_class, max_iterations):
    
    weights = np.ones(num_class)/num_class
    v_values,ll = maximization_step(weights,data,collections,offer_times)
    #print('initial loglikelihood values',ll)
    #print('intitial preference parameters',v_values)

    for i in range(max_iterations):
        curr_ll = ll
        weights,ll = expectation_step(v_values,data,collections,num_class,offer_times)
        #print(f'the loglikelihood after expectation at iteration {i} is {ll}')
        v_values,ll = maximization_step(weights,data,collections,offer_times)
        #print(f'the loglikelihood after maximization at iteration {i} is {ll}')
        
        ''' if ll - curr_ll < 0.001:
            print(f'{ll} at current iteration and {curr_ll} at last iteration  ')
            return weights, v_values,ll '''
        
    return   weights, v_values,ll
        
```

```python
def collection_distribution_lcmnl2(n_num,collection,num_mix,weights,parameters_v):
    
    
    collection_distribution = np.zeros((n_num,len(collection)))
    
    for i in range(len(collection)): 
        curr_total_v = np.zeros(num_mix) # the denominator for each class
        
        for j in range(len(collection[i])): 
            for k in range(num_mix):
                curr_total_v[k] = curr_total_v[k] + math.exp(parameters_v[k][collection[i][j]])
        #print('checking the weights of denominators',curr_total_v)
        
        for j in range(len(collection[i])):
            for k in range(num_mix):
                collection_distribution[collection[i][j]][i] = collection_distribution[collection[i][j]][i] + weights[k]*math.exp(parameters_v[k][collection[i][j]])/curr_total_v[k]
    
    return collection_distribution
```

```python
def deterministic_term(data,offertimes):
    dt = 0 
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            if data[i][j] >0:
                dt = dt + offertimes[j]*data[i][j]*np.log(data[i][j])
    return dt,dt/np.sum(offertimes)
```

```python
## record the limit loss of all instances
# in average 
all_lcmnl2_limit_loss = []
all_lcmnl2_limit_prob = []
all_lcmnl2_limit_runtime = []

# EM algorithm parameters
max_iterations = 50
num_class = 2

for i in range(len(pred_offer_times_list)):
    

    ## define container for lcmnl2 limit 
    
    lcmnl2_limit_loss_collection =[]
    lcmnl2_limit_runtime_collection =[]
    lcmnl2_limit_probability_collection =[]
    
    for j in range(pred_instance_size[i]):
        print(f'testing instance with offertimes {pred_offer_times_list[i]} : {j}th ')
    
        starttime = time.time()
        weights, v_values,ll = expectation_maximization(all_full_instances[i][j],all_full_collections[i],all_full_offertimes[i], num_class, max_iterations)
        lcmnl2_runtime = time.time()-starttime
        
        lcmnl2_limit_prob = collection_distribution_lcmnl2(n-1,all_full_collections[i],num_class,weights,v_values)
        curr_deterministic_term, curr_avg_deterministic_term = deterministic_term(all_full_instances[i][j],all_full_offertimes[i])
        
        lcmnl2_limit_loss_collection.append(-ll/np.sum(all_full_offertimes[i]) + curr_avg_deterministic_term)
        lcmnl2_limit_probability_collection.append(lcmnl2_limit_prob)
        lcmnl2_limit_runtime_collection.append(lcmnl2_runtime)
        
        print('lcmnl2 kl loss is', lcmnl2_limit_loss_collection[-1])
        print('lcmnl2 kl runtime is', lcmnl2_runtime)
    
       
    ## for each train instance [i][j] limit loss result is a number
    all_lcmnl2_limit_loss.append(lcmnl2_limit_loss_collection)
    all_lcmnl2_limit_runtime.append(lcmnl2_limit_runtime_collection)
    ## for each train instance [i][j] limit probability is a matrix
    all_lcmnl2_limit_prob.append(lcmnl2_limit_probability_collection)
```

```python
def create_new_list(A, error_i, error_j):
    # Create a set of (i, j) pairs for fast lookup of error positions
    error_set = set(zip(error_i, error_j))
    
    # Create a new list of lists by excluding the elements that are in the error_set
    A_prime = []
    for i, row in enumerate(A):
        new_row = [elem for j, elem in enumerate(row) if (i, j) not in error_set]
        A_prime.append(new_row)
    
    return A_prime

error_instance_df = pd.read_csv('mccm_error_instance.csv')
error_i_raw = error_instance_df['error_i'].tolist()
error_j_raw = error_instance_df['error_j'].tolist()

error_i = []
error_j = []
for x in range(len(error_i_raw)):
    if error_i_raw[x]<=1:
        error_i.append(error_i_raw[x])
        error_j.append(error_j_raw[x])

new_all_lcmnl2_loss = create_new_list(all_lcmnl2_limit_loss,error_i,error_j)
new_all_lcmnl2_time = create_new_list(all_lcmnl2_limit_runtime,error_i,error_j)
new_all_lcmnl2_prob = create_new_list(all_lcmnl2_limit_prob,error_i,error_j)

for i in range(len(pred_offer_times_list)):
    df_lcmnl2_loss = pd.DataFrame({'lcmnl2_kl_loss':new_all_lcmnl2_loss[i],'lcmnl2_kl_time':new_all_lcmnl2_time[i]})
    df_lcmnl2_loss.to_csv('lcmnl_2class/limit_loss_new/'+str(pred_offer_times_list[i])+'.csv')

lcmnl2_kl_loss_mean = [np.mean(new_all_lcmnl2_loss[x]) for x in range(len(pred_offer_times_list)) ]
lcmnl2_kl_loss_se = [ np.std(new_all_lcmnl2_loss[x])/np.sqrt(len(new_all_lcmnl2_loss[x])) for x in range(len(pred_offer_times_list)) ]

lcmnl2_kl_time_mean = [np.mean(new_all_lcmnl2_time[x]) for x in range(len(pred_offer_times_list)) ]
lcmnl2_kl_time_se = [ np.std(new_all_lcmnl2_time[x])/np.sqrt(len(new_all_lcmnl2_time[x])) for x in range(len(pred_offer_times_list)) ]

df_lcmnl2_kl_loss = pd.DataFrame({'offertimes':pred_offer_times_list,'collection_size':full_collection_size,
                                  'lcmnl2_kl_loss':lcmnl2_kl_loss_mean, 'lcmnl2_kl_loss_se':lcmnl2_kl_loss_se,
                                'lcmnl2_kl_time':lcmnl2_kl_time_mean, 'lcmnl2_kl_time_se':lcmnl2_kl_time_se
                                  })
df_lcmnl2_kl_loss.to_csv('lcmnl_2class/lcmnl2_kl_summary_NEW.csv')
print(df_lcmnl2_kl_loss)

all_rep = []
rep_mean = []
for i in range(len(pred_instance_size)):
    rep_lst = []
    for j in range(len(new_all_lcmnl2_loss[i])):
        if new_all_lcmnl2_loss[i][j] <1e-6:
            rep_lst.append(j)
    df_rep = pd.DataFrame(rep_lst)
    df_rep.to_csv('lcmnl_2class/representability_new/'+str(pred_offer_times_list[i])+'.csv')
    all_rep.append(rep_lst)
    rep_mean.append(len(rep_lst)/len(new_all_lcmnl2_loss[i]))
print(rep_mean)

df_rep = pd.DataFrame({'offertimes':pred_offer_times_list,'collection_size':full_collection_size,'lcmnl2_rep':rep_mean})
df_rep.to_csv('lcmnl_2class/lcmnl2_rep_summary_NEW.csv')
print(df_rep)
```
