```python
## compute standard error
offertimes = [20,30,40,50,60]
mdm_ci_lb_reg_se_lst = []
mdm_ci_ub_reg_se_lst = []
mdm_etp_lb_reg_se_lst = []
mdm_etp_ub_reg_se_lst = []

mdm_l1ci_reg_se_lst = []

rum_ci_reg_se_lst = []
rum_etp_reg_se_lst = []

mccm_reg_se_lst = []
lcmnl_reg_se_lst = []
mnl_reg_se_lst = []
mem_reg_se_lst = []


for times in offertimes:
    df_mdm_ci = pd.read_csv('prediction/mdm/revenue_prediction/details1/'+str(times)+'.csv')
    df_mdm_ci_lb_reg = df_mdm_ci['mdm_lb_best_rev_diff']
    df_mdm_ci_ub_reg = df_mdm_ci['mdm_ub_best_rev_diff']
    mdm_ci_lb_reg = np.array(df_mdm_ci_lb_reg)
    mdm_ci_lb_reg_se = np.std(mdm_ci_lb_reg)/np.sqrt(len(mdm_ci_lb_reg))
   
    mdm_ci_ub_reg = np.array(df_mdm_ci_ub_reg)
    mdm_ci_ub_reg_se = np.std(mdm_ci_ub_reg)/np.sqrt(len(mdm_ci_ub_reg))
   
    mdm_ci_lb_reg_se_lst.append(mdm_ci_lb_reg_se)
    mdm_ci_ub_reg_se_lst.append(mdm_ci_ub_reg_se)
    
    
    df_mdm_etp = pd.read_csv('prediction/etp_mdm/revenue_prediction/details1/'+str(times)+'.csv')
    df_mdm_etp_lb_reg = df_mdm_etp['mdm_lb_best_rev_diff']
    df_mdm_etp_ub_reg = df_mdm_etp['mdm_ub_best_rev_diff']
    mdm_etp_lb_reg = np.array(df_mdm_etp_lb_reg)
    mdm_etp_lb_reg_se = np.std(mdm_etp_lb_reg)/np.sqrt(len(mdm_etp_lb_reg))
    
    mdm_etp_ub_reg = np.array(df_mdm_etp_ub_reg)
    mdm_etp_ub_reg_se = np.std(mdm_etp_ub_reg)/np.sqrt(len(mdm_etp_ub_reg))
    
    mdm_etp_lb_reg_se_lst.append(mdm_etp_lb_reg_se)
    mdm_etp_ub_reg_se_lst.append(mdm_etp_ub_reg_se)
    
    df_rum_ci = pd.read_csv('prediction/rum/revenue_prediction/details1/'+str(times)+'.csv')
    df_rum_ci_reg = df_rum_ci['rum_lb_best_rev_diff']
    rum_ci_reg = np.array(df_rum_ci_reg)
    rum_ci_reg_se = np.std(rum_ci_reg)/np.sqrt(len(rum_ci_reg))
    rum_ci_reg_se_lst.append(rum_ci_reg_se)

    
    df_mem = pd.read_csv('prediction/mem/revenue_prediction/details1/'+str(times)+'.csv')
    df_mem_reg = df_mem['mem_best_rev_diff']
    mem_reg = np.array(df_mem_reg)
    mem_reg_se = np.std(mem_reg)/np.sqrt(len(mem_reg)) ### standard deviation/\sqrt(number of samples)
    mem_reg_se_lst.append(mem_reg_se)

    df_mnl = pd.read_csv('prediction/mnl/revenue_prediction/details1/'+str(times)+'.csv')
    df_mnl_reg = df_mnl['mnl_best_rev_diff']
    mnl_reg = np.array(df_mnl_reg)
    mnl_reg_se = np.std(mnl_reg)/np.sqrt(len(mnl_reg))
    mnl_reg_se_lst.append(mnl_reg_se)
        
    df_mccm = pd.read_csv('prediction/mccm/revenue_prediction/details1/'+str(times)+'.csv')
    df_mccm_reg = df_mccm['mccm_best_rev_diff']
    mccm_reg = np.array(df_mccm_reg)
    mccm_reg_se = np.std(mccm_reg)/np.sqrt(len(mccm_reg))
    mccm_reg_se_lst.append(mccm_reg_se)
    
     
    df_lcmnl = pd.read_csv('prediction/lcmnl/revenue_prediction/details1/'+str(times)+'.csv')
    df_lcmnl_reg = df_lcmnl['lcmnl_best_rev_diff']
    lcmnl_reg = np.array(df_lcmnl_reg)
    lcmnl_reg_se = np.std(lcmnl_reg)/np.sqrt(len(lcmnl_reg))
    lcmnl_reg_se_lst.append(lcmnl_reg_se)   
```
