```python
import pandas as pd
import numpy as np
```

```python
df_mdm = pd.read_csv('limit/mdm/mdm_limit_summary.csv')
df_mdm = df_mdm.drop(df_mdm.columns[0],axis=1)
df_mdm = df_mdm.rename(columns={'avg_mdm_limit_runtime': 'mdm_runtime'})
df_mdm
```

```python
df_rum = pd.read_csv('limit/rum/rum_limit_summary.csv')
df_rum = df_rum.drop(df_rum.columns[0],axis=1)
df_rum = df_rum.rename(columns={'avg_rum_limit_runtime': 'rum_runtime'})
df_rum
```

```python
df_mem = pd.read_csv('limit/mem/mem_limit_summary.csv')
df_mem = df_mem.drop(df_mem.columns[0],axis=1)
df_mem = df_mem.rename(columns={'avg_mem_limit_runtime': 'mem_runtime'})
df_mem
```

```python
df_mnl = pd.read_csv('limit/mnl/mnl_limit_summary.csv')
df_mnl = df_mnl.drop(df_mnl.columns[0],axis=1)
df_mnl = df_mnl.rename(columns={'avg_mnl_limit_runtime': 'mnl_runtime'})
df_mnl
```

```python
df_mccm = pd.read_csv('limit/mccm/mccm_limit_summary.csv')
df_mccm = df_mccm.drop(df_mccm.columns[0],axis=1)
df_mccm = df_mccm.rename(columns={'avg_mccm_limit_runtime': 'mccm_runtime'})
df_mccm
```

```python
df_lcmnl = pd.read_csv('limit/lcmnl/lcmnl_limit_summary.csv')
df_lcmnl = df_lcmnl.drop(df_lcmnl.columns[0],axis=1)
df_lcmnl = df_lcmnl.rename(columns={'avg_lcmnl_limit_runtime': 'lcmnl_runtime'})
df_lcmnl
```

```python
limit_result = pd.concat([df_mdm['offertimes'],
                          df_mdm['total_mdm_loss'],df_mdm['total_mdm_loss_se'],df_mdm['mdm_runtime'],
                          df_rum['total_rum_loss'],df_rum['total_rum_loss_se'],df_rum['rum_runtime'],
                          df_mem['total_mem_loss'],df_mem['total_mem_loss_se'],df_mem['mem_runtime'],
                          df_mnl['total_mnl_loss'],df_mnl['total_mnl_loss_se'],df_mnl['mnl_runtime'],
                          df_mccm['total_mccm_loss'],df_mccm['total_mccm_loss_se'],df_mccm['mccm_runtime'],
                          df_lcmnl['total_lcmnl_loss'],df_lcmnl['total_lcmnl_loss_se'],df_lcmnl['lcmnl_runtime']], axis=1)
limit_result.to_csv('limit/summary/total_loss_summary.csv')
limit_result
```

```python
avg_loss_result = pd.concat([df_mdm['offertimes'],
                          df_mdm['avg_mdm_loss'],df_mdm['avg_mdm_loss_se'],df_mdm['mdm_runtime'],
                          df_rum['avg_rum_loss'],df_rum['avg_rum_loss_se'],df_rum['rum_runtime'],
                          df_mem['avg_mem_loss'],df_mem['avg_mem_loss_se'],df_mem['mem_runtime'],
                          df_mnl['avg_mnl_loss'],df_mnl['avg_mnl_loss_se'],df_mnl['mnl_runtime'],
                          df_mccm['avg_mccm_loss'],df_mccm['avg_mccm_loss_se'],df_mccm['mccm_runtime'],
                          df_lcmnl['avg_lcmnl_loss'],df_lcmnl['avg_lcmnl_loss_se'],df_lcmnl['lcmnl_runtime']], axis=1)
avg_loss_result.to_csv('limit/summary/average_loss_summary.csv')
avg_loss_result
```
